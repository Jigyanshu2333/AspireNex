```python
data['Contract'] = data['Contract'].replace(['Month-to-month','Two year','One year'],[0,1,2])
```

```python
data['PaymentMethod'] = data['PaymentMethod'].replace(['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'],[0,1,2,3])
```

```python
sns.kdeplot(data.MonthlyCharges[(data['Churn'] == "Yes")],color = "Red",shade =  True)
sns.kdeplot(data.MonthlyCharges[(data['Churn'] == "No")],color = "Blue",shade =  True)
plt.legend(['Churn','Not Churn'])
plt.show()
```

```python
data['TotalCharges'] = data['TotalCharges'].replace(' ',None)
```

```python
data.dropna(inplace = True)
```

```python
data['TotalCharges'].dtypes
```

```python
data['TotalCharges'] = data['TotalCharges'].astype(float)
data.info()
```

```python
sns.kdeplot(data.TotalCharges[(data['Churn'] == "Yes")],color = "Red",shade =  True)
sns.kdeplot(data.TotalCharges[(data['Churn'] == "No")],color = "Blue",shade =  True)
plt.legend(['Churn','Not Churn'])
plt.show()
```

```python
data[['PaperlessBilling','Churn']] = data[['PaperlessBilling','Churn']].replace(['No','Yes'],[0,1])
```

```python
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
data[['tenure','MonthlyCharges','TotalCharges']] = scale.fit_transform(data[['tenure','MonthlyCharges','TotalCharges']])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
```

```python
x = data.drop(['Churn'],axis = 1)
y = data['Churn']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,stratify = data['Churn'])
```

```python
LRModel = LogisticRegression()
LRModel.fit(x_train,y_train)
y_pred = LRModel.predict(x_test)
print('Accuracy of the mode is: '+ str(accuracy_score(y_pred,y_test)*100))
```

```python
RFModel = RandomForestClassifier()
RFModel.fit(x_train,y_train)
y_pred = RFModel.predict(x_test)
print('Accuracy of the mode is: '+ str(accuracy_score(y_pred,y_test)*100))
```

```python
GDModel = SGDClassifier()
GDModel.fit(x_train,y_train)
y_pred = GDModel.predict(x_test)
print('Accuracy of the mode is: '+ str(accuracy_score(y_pred,y_test)*100))
```
